```python
#import tsla.csv using pandas
import pandas as pd
df = pd.read_csv('tsla.csv')
```

```python
# show data using .info()
df.info()
```

```python
# show data using .head()
df.head()
```

```python
# create bar grahp using .bar using plotly x = date, y = volume
import plotly.express as px
import plotly.graph_objs as go

fig = go.Figure(data=[go.Bar(x=df['Date'], y=df['Volume'])])
fig.show()
```

```python
# create bar grahp using .bar using plotly x = date, y = volume but the color of the bars changes depending on the value of the volume
fig = go.Figure(data=[go.Bar(x=df['Date'], y=df['Volume'], marker_color=df['Volume'])])
fig.show()
```

```python
# add to previous graph a black stroke around the bars with with 2 pixel
fig = go.Figure(data=[go.Bar(x=df['Date'], y=df['Volume'], marker_color=df['Volume'], marker_line_color='rgb(0,0,0)', marker_line_width=2)])
fig.show()
```

```python
# show header for the previous diagramm on middle top of the graph 
fig.update_layout(title_text='TSLA Volume', title_x=0.5, title_font_size=20)
```

```python
# add a title to the x axis and y axis with font = 16 and for x axis rotate the labels by 315 degrees
fig.update_xaxes(title_text='Date', title_font_size=16, tickangle=315)
fig.update_yaxes(title_text='Volume', title_font_size=16)
```

```python
# for x and y axis set the lables font = 14
fig.update_xaxes(tickfont_size=14)
fig.update_yaxes(tickfont_size=14)
```

```python
# make the graph width for all workspace and make the highth 700 pixel
fig.update_layout(width=None, height=700)
```

```python
# for the previous graph remove extra padding around the edges
fig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
```

```python
# make a pie chart but need to use all parameters from the previous bar graph use piefig isntead of fig
piefig = go.Figure(data=[go.Pie(labels=df['Date'], values=df['Volume'])])
piefig.show()
```

```python
# make the pie chart with the same parameters as the bar graph
piefig = go.Figure(data=[go.Pie(labels=df['Date'], values=df['Volume'], marker_colors=df['Volume'], marker_line_color='rgb(0,0,0)', marker_line_width=2)])
piefig.update_layout(title_text='TSLA Volume', title_x=0.5, title_font_size=20)
piefig.update_layout(margin=dict(l=0, r=0, b=0, t=0))
piefig.update_layout(width=None, height=700)
piefig.update_xaxes(title_text='Date', tickfont_size=14, title_font_size=16, tickangle=315)
piefig.update_yaxes(title_text='Volume', tickfont_size=14, title_font_size=16)
piefig.show()
```

```python
# create a line graph x = date, y = volume 
linefig = go.Figure(data=[go.Scatter(x=df['Date'], y=df['Volume'])])
linefig.show()
```

```python
# make the graph with crimson lines and white dots, color of the dots border is black with 1 pixels width
linefig = go.Figure(data=[go.Scatter(x=df['Date'], y=df['Volume'], mode='lines+markers', line_color='crimson', marker_color='white', marker_line_color='black', marker_line_width=1)])
linefig.show()
```

```python
# add ivory grid to the graph with 2 pixels width
linefig.update_layout(plot_bgcolor='ivory', xaxis_gridwidth=2, yaxis_gridwidth=2)
```
